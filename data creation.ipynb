```python
fake = Faker()
cities = ['London', 'Paris', 'Berlin', 'Rome', 'Madrid', 'Vienna', 'Dublin', 'Lisbon', 'Warsaw', 'Amsterdam']
streets = ['Baker Street', 'Fleet Street', 'Oxford Street', 'Wall Street', 'Abbey Road', 'Fifth Avenue', 'Elm Street', 'High Street', 'Park Avenue', 'Main Street']
```

```python
for i in range(len(shop_list)):
    id_part = shop_list[i]
    city = cities[randint(0, 9)]
    street_name = streets[randint(0, 9)]
    street_number = fake.building_number()
    print(f"{id_part}\t{city}, {street_name}, {street_number}, {id_part}")
```

```python
with open('synthetic_data_shops.csv', 'w', newline='') as csvfile:
    fieldnames = ['ID', 'Address']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(len(shop_list)):
        id_part = shop_list[i]
        city = cities[randint(0, 9)]
        street_name = streets[randint(0, 9)]
        street_number = fake.building_number()
        writer.writerow({'ID': id_part, 'Address': f"{city}, {street_name} {street_number}"})
```

```python
category_1 = ['Skin Care', 'Hair Care', 'Makeup', 'Fragrance', 'Bath & Body']
category_2 = ['Cleansers', 'Moisturizers', 'Masks', 'Shampoos', 'Conditioners', 'Lipstick', 'Mascara', 'Perfume', 'Body Wash', 'Body Lotion']
category_3 = ['Face Cleansers', 'Body Cleansers', 'Day Creams', 'Night Creams', 'Face Masks', 'Hair Masks', 'Matte Lipstick', 'Waterproof Mascara', 'Eau de Parfum', 'Body Scrubs']
brands = ['L’Oréal', 'Unilever', 'Estée Lauder', 'Procter & Gamble', 'Coty']

data = []

for i in range(200):  # generate 1000 products
    product_id = i + randint(1000, 3000)
    product_name = f"{brands[randint(0, 4)]} {fake.bs()} {randint(1, 500)}ml"
    data.append({
        'id': product_id,
        'product_name': product_name,
        'category_1': category_1[randint(0, 4)],
        'category_2': category_2[randint(0, 9)],
        'category_3': category_3[randint(0, 9)],
        'brand': brands[randint(0, 4)]
    })

df = pd.DataFrame(data)

# write DataFrame to CSV
df.to_csv('product_data.csv', index=False, encoding='utf-8')
```

```python
# load the shop_id from the previously created CSV
df_product = pd.read_csv('shops.csv')
shop_ids = df_product['id'].unique().tolist()

data = []

for i in range(1000):  # generate 1000 transactions
    id = fake.unique.random_number(digits=5)
    user_id = fake.random_int(min=1000, max=9999)
    transaction_type = np.random.choice([0, 1], p=[0.95, 0.05])  # 90% probability for 0 and 10% for 1
    
    # generate purchase_date with random time component
    purchase_date = fake.date_time_between(start_date='-90d', end_date='-60d')
    
    # processed_at is same as purchase_date
    processed_at = purchase_date.date()
    
    gross_transaction_amount = randint(100, 10000)
    discount_amount = randint(0, 100)
    net_transaction_amount = gross_transaction_amount - discount_amount
    shop_id = np.random.choice(shop_ids)  # randomly choose a shop_id from the list

    data.append({
        'id': id,
        'user_id': user_id,
        'transaction_type': transaction_type,
        'purchase_date': purchase_date,
        'processed_at': processed_at,
        'gross_transaction_amount': gross_transaction_amount,
        'discount_amount': discount_amount,
        'net_transaction_amount': net_transaction_amount,
        'shop_id': shop_id
    })

df = pd.DataFrame(data)

# write DataFrame to CSV
df.to_csv('transaction_data.csv', index=False, encoding='utf-8')
```

```python
df_product = pd.read_csv('product_data.csv')
product_ids = df_product['id'].unique().tolist()

df_transactions = pd.read_csv('transaction_data.csv')
purchase_ids = df_transactions['id'].unique().tolist()
```

```python
data = []

for purchase_id in purchase_ids:
    num_items = randint(1, 10)  # random number of items between 1 and 10
    for _ in range(num_items):
        product_id = np.random.choice(product_ids)
        id = f'{purchase_id}-{_}'
        amount = np.random.choice([1, 2, 3], p=[0.7, 0.2, 0.1])
        price = randint(100, 1000)
        discount = randint(0, 100)
        marketing_rule = np.random.choice([0, 1])

        data.append({
            'id': id,
            'purchase_id': purchase_id,
            'product_id': product_id,
            'amount': amount,
            'price': price,
            'discount': discount,
            'marketing_rule': marketing_rule
        })

df = pd.DataFrame(data)

# write DataFrame to CSV
df.to_csv('items_data.csv', index=False, encoding='utf-8')
```
